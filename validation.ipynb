Function to get sleep times for each goose from solomon

```python
def valid_solo(sipath,fname,opath):
    index_lst = []
    lrow_lst = []
    rrow_lst = []
    for i in fname:
        Dataframe = pd.read_csv(sipath+i)
        # Dataframe = pd.read_csv(r'L:\DLCValidation\20220725\csv_solomon\bs6165_5downsampled Goose31 .csv')
        # Dataframe = pd.read_csv(r'C:\Users\labgrprattenborg\Desktop\Pradeep\20220725_DLC_extract\6165\bs6165_05downsampledcro_goose38.csv')
        Dataframe = Dataframe.replace( to_replace= "Right Open", value=2)
        Dataframe = Dataframe.replace( to_replace= "Left Open", value=2)
        Dataframe = Dataframe.replace( to_replace= "Right Close", value=3)
        Dataframe = Dataframe.replace( to_replace= "Left Close", value=3)
        Dataframe = Dataframe.replace( to_replace= "Right Out of Sight", value=1)
        Dataframe = Dataframe.replace( to_replace= "Left Out of Sight", value=1)
        Dataframe = Dataframe.replace( to_replace= "Not sure", value=0)

        k = Dataframe[Dataframe['L_Goose31'] > 2].shape[0]
        l = Dataframe[Dataframe['R_Goose31'] > 2].shape[0]
        m = k+l
        
        index_lst.append(i)
        lrow_lst.append(m)
        # rrow_lst.append(l)

    df = pd.DataFrame(columns=['File_name', 'Sol_sleep_row'])
    df['File_name'] = index_lst
    df['Sol_sleep_row'] = lrow_lst
    # df['Sol_sleep_Right'] = rrow_lst
    df.to_csv(opath+fname[-1]+'_{}.csv'.format('solo'))
```

Function to get sleep times for each goose from DLC

```python
def valid_dlc(dipath,fname,opath):
    index_lst = []
    lrow_lst = []
    for i in fname:
        Dataframe = pd.read_csv(dipath+i)
        # Dataframe = pd.read_csv(r'L:\DLCValidation\20220725\csv_solomon\bs6165_5downsampled Goose31 .csv')
        # Dataframe = pd.read_csv(r'C:\Users\labgrprattenborg\Desktop\Pradeep\20220725_DLC_extract\6165\bs6165_05downsampledcro_goose38.csv')
        Dataframe = Dataframe.replace( to_replace= "Right Open", value=2)
        Dataframe = Dataframe.replace( to_replace= "Left Open", value=2)
        Dataframe = Dataframe.replace( to_replace= "Right Close", value=3)
        Dataframe = Dataframe.replace( to_replace= "Left Close", value=3)
        Dataframe = Dataframe.replace( to_replace= "Right Out of Sight", value=1)
        Dataframe = Dataframe.replace( to_replace= "Left Out of Sight", value=1)
        Dataframe = Dataframe.replace( to_replace= "Not sure", value=0)

        k = Dataframe[Dataframe['Right_eye'] > 2].shape[0]
        l = Dataframe[Dataframe['Left_eye'] > 2].shape[0]
        m = k+l
        
        index_lst.append(i)
        lrow_lst.append(m)
        # rrow_lst.append(l)

    df = pd.DataFrame(columns=['File_name', 'Dlc_sleep_row'])
    df['File_name'] = index_lst
    df['Dlc_sleep_row'] = lrow_lst
    # df['Sol_sleep_Right'] = rrow_lst
    df.to_csv(opath+fname[-1]+'_{}.csv'.format('dlc'))
```
